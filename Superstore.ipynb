```python
import pandas as pd 
```

```python
df=pd.read_csv("Sample - Superstore.csv", encoding='latin1')
```

```python
df
```

```python
df.head(10)
```

```python
df.info()
```

```python
data=df.drop_duplicates()
```

```python
data
```

```python
data["Order Date"]=pd.to_datetime(data["Order Date"],errors="coerce")
```

```python
data.info()
```

```python
data["Order Date"].isnull().sum
```

```python
data["Ship Date"]=pd.to_datetime(data["Ship Date"],errors="coerce")
```

```python
data["Ship Date"].isnull().sum
```

```python
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['Day'] = data['Order Date'].dt.day
```

```python
data['Profit Margin'] = data['Profit'] / data['Sales']
```

```python
data.head(4)
```

```python
data['Shipping Days'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
monthly = data.groupby(data['Order Date'].dt.to_period('M'))[['Sales','Profit']].sum().reset_index()
```

```python
data.info()
```

```python
monthly
```

```python
category = data.groupby('Category')[['Sales','Profit']].sum()

sub_category = data.groupby('Sub-Category')[['Sales','Profit']].sum()
```

```python
category
```

```python
sub_category
```

```python
region=data.groupby("Region")[["Sales","Profit"]].sum()
```

```python
top_customers = data.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
```
